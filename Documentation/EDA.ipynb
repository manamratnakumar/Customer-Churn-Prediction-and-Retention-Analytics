What is the overall customer churn rate?

Business Objective:

Understand the percentage of customers who have churned versus retained.

```python
churn_count = df['Churn'].value_counts()
print(churn_count)

churn_percentage = (df['Churn'].value_counts(normalize=True) * 100).round(2)
print(churn_percentage)
```

```python
df['Churn'].value_counts().plot(kind='bar')

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

Business Insight:
The organization should monitor the churn rate as a key business KPI. A higher churn rate directly impacts customer retention and revenue.

Identify relationships among numerical variables.

```python
corr = df[['tenure',
           'MonthlyCharges',
           'TotalCharges',
           'numAdminTickets',
           'numTechTickets']].corr()

corr
```

```python
plt.figure(figsize=(8,6))

plt.imshow(corr, cmap='coolwarm')

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
